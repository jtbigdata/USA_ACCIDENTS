```python
#===================================callback1============================================
@app.callback(
        Output(component_id='graph_1', component_property='figure'),
        [Input(component_id='dropdown_1', component_property='value'),Input(component_id='dropdown_2', component_property='value')
         ])

def update_graph(dropdown_1,dropdown_2):
    df1 = df[['Año','Mes','Dia_semana','Hora','Periodo','Provincia', 'Condado', 'Ciudad', 'Calle','Severidad']]
    df1=df1[df1['Periodo'].isin(dropdown_2)]
    frecuencia = df1.groupby(['Provincia', 'Severidad']).size().reset_index(name='count')
    frecuencia['percentage'] = frecuencia['count'] / frecuencia.groupby('Provincia')['count'].transform('sum') * 100

    fig1 = px.bar(frecuencia, x='Provincia', y="percentage", color="Severidad", barmode="group",
                  color_continuous_scale='jet'
                  )
    fig1.update_layout(
        plot_bgcolor='black',paper_bgcolor='black',
        title=dict(text="Porcentaje de ocurrencia vs Provincia.", x=0.5, font=dict(color="ORANGE")),
        xaxis=dict(title='Provincia',tickfont=dict(color='white'),title_font=dict(color='white')),
        yaxis=dict(title='Porcentaje de ocurrencia (%)',tickfont=dict(color='white'),title_font=dict(color='white')) ,
        font=dict(color="white"),coloraxis_colorbar=dict(title=dict(text="Severidad", font=dict(color='white')))
    )
    return fig1
#===================================callback2=============================================
@app.callback(
        Output(component_id='graph_2', component_property='figure'),
        [Input(component_id='dropdown_1', component_property='value'),
        Input(component_id='dropdown_2', component_property='value')
         ]   )
def update_graph(dropdown_1,dropdown_2):

    df1 = df[['Año','Mes','Dia_semana','Hora','Periodo',
              'Provincia','Condado','Ciudad','Calle','Severidad']]

    df1=df1[df1['Periodo'].isin(dropdown_2)]
    df2 = df1[[dropdown_1,'Provincia']].value_counts().reset_index(name='count')
    df2['percentage'] = df2.groupby(dropdown_1)['count'].transform(lambda x: (x / x.sum()) * 100)

    valores_ordenados = sorted(df2[dropdown_1].unique())

    fig2 = px.scatter(df2, x=dropdown_1, y='percentage',color='Provincia',
                color_discrete_sequence=px.colors.qualitative.Dark24)

    fig2.update_traces(marker=dict(size=14,
                               symbol='circle'
                        ))

    fig2.update_layout(
                    paper_bgcolor='black',plot_bgcolor='black',
                    xaxis=dict(title='',tickmode='linear', nticks=len(valores_ordenados),tickfont=dict(color='white'),title_font=dict(color='white')),
                    title=dict(text="Porcentaje de ocurencia vs " + str(dropdown_1) +'.', x=0.5, font=dict(color="ORANGE")),
                    yaxis=dict(title='Porcentaje de ocurrencia (%)',tickfont=dict(color='white'),title_font=dict(color='white')),
                    font=dict(color="white"),
                    coloraxis_colorbar=dict(title=dict(text="Severidad", font=dict(color='white')))
                        )
    return fig2

#===================================callback3=============================================
@app.callback(
        Output(component_id='graph_3', component_property='figure'),
        [Input(component_id='dropdown_f21', component_property='value'),
        Input(component_id='dropdown_f22', component_property='value'),
        Input(component_id='dropdown_f24', component_property='value'),
         ]   )


def update_graph(dropdown_f21,dropdown_f22,dropdown_f24):
    df1 = df[['Año','Mes','Dia_semana','Hora','Periodo','Provincia',
              #--------------------------------------------------------
              'Congelación', 'Despejado', 'Despejado/Ventoso', 'Granizo',
              'Humo','Humo/Ventoso', 'Lluvia', 'Lluvia Intensa',
              'Lluvia Intensa/Ventoso','Lluvia Ligera', 'Lluvia Ligera/Congelación',
              'Lluvia Ligera/Truenos','Lluvia Ligera/Ventoso', 'Lluvia/Ventoso',
              'Mezcla Invernal', 'Neblina','Neblina Ligera', 'Neblina Ligera/Congelación',
              'Neblina/Ventoso','Nieve','Nieve Intensa','Nieve Intensa/Ventoso',
              'Nieve Ligera','Nieve Ligera/Ventoso','Nieve/Ventoso','Nublado',
              'Nublado/Ventoso','Polvo','Polvo/Ventoso','Tormentas',
              'Tormentas Intensas','Tormentas Intensas/Ventoso','Tormentas/Lluvia',
              'Tormentas/Lluvia Intensa', 'Tormentas/Ventoso']
              #---------------------------------------------------------------
              ]


    df1=df1[df1['Provincia'].isin([dropdown_f24])]
    valores_ordenados = sorted(df1[dropdown_f21].unique())


    fig = px.density_heatmap(df1, x=dropdown_f21, y=dropdown_f22 ,color_continuous_scale='jet')
    fig.update_layout( plot_bgcolor='black',
                       paper_bgcolor='black',
                       title=dict(text= str(dropdown_f22)+ '  vs ' +str(dropdown_f21)+'.', x=0.5,font=dict(color="ORANGE")),
                       xaxis=dict(tickmode='linear', nticks=len(valores_ordenados),tickfont=dict(color='white'),title_font=dict(color='white')),
                       yaxis=dict(
                        tickmode='array',tickvals=[0, 1],ticktext=['0', '1'],
                        title=dropdown_f22,tickfont=dict(color='white'),title_font=dict(color='white')),
                       font=dict(color="white"),coloraxis_colorbar=dict(title=dict(text="Severidad", font=dict(color='white')))
                    )

    return fig

#===================================callback4=============================================
@app.callback(
        Output(component_id='graph_4', component_property='figure'),
        [
        Input(component_id='dropdown_f24', component_property='value'),
        Input(component_id='dropdown_f25', component_property='value')
         ]   )

def update_graph(dropdown_f24,dropdown_f25):
    df2 = df[['Provincia', dropdown_f25]].value_counts().reset_index()
    df2 = df2.rename(columns={0: 'count'})
    xa = df2.loc[df2['Provincia'] == dropdown_f24]


    fig2= px.pie(xa, values='count', names=dropdown_f25,
                 color_discrete_sequence=px.colors.qualitative.Dark24
                 )
    fig2.update_layout(plot_bgcolor='black',
                       paper_bgcolor='black',
                    title=dict(text="Porcentaje de ocurencia vs " +
                               str(dropdown_f25)+'.', x=0.5,font=dict(color="ORANGE")),
                    yaxis=dict(title=''),
                    font=dict(color="white"),
                    coloraxis_colorbar=dict(title=dict(text="Severidad", font=dict(color='white')))
                        )

    return fig2



#===================================callback5=============================================
@app.callback(
        Output(component_id='graph_5', component_property='figure'),
        Input(component_id='dropdown_6', component_property='value'),
        Input(component_id='checkllist', component_property='value')
            )
def update_graph(dropdown_6,checkllist):

    columnas_seleccionadas = ['Severidad','Calle','Ciudad','Provincia','Lat_i', 'Lng_i', dropdown_6,'Periodo','Condicion_clima']
    df1 = df[columnas_seleccionadas]
    df1=df1[df1['Provincia']==checkllist]
    df1 = df1.sort_values(dropdown_6)
    conteo_ciudad = df1['Ciudad'].value_counts().head(10)
    top_ciudades = conteo_ciudad.nlargest(10).index.tolist()

    fig = px.scatter_mapbox(df1, lat="Lat_i", lon="Lng_i",
                            color="Severidad", size="Severidad",
                            mapbox_style='open-street-map',
                            color_continuous_scale='jet',
                            #color_continuous_scale=px.colors.cyclical.IceFire,
                            size_max=8, zoom=4,
                            animation_frame=dropdown_6,  # O puedes usar 'Hora' si prefieres
                            #title='Evolución Temporal',
                            labels={'Severidad': 'Nivel de Severidad'},
                            range_color=[min(df1['Severidad']), max(df1['Severidad'])],
                            hover_data={'Calle': True, 'Ciudad': True, 'Severidad': True,'Periodo':True,
                                        'Condicion_clima':True,'Lat_i':False, 'Lng_i':False}
                            )
    fig.update_layout(plot_bgcolor='black',paper_bgcolor='black',
                      mapbox=dict(center={'lat': df1['Lat_i'].mean(), 'lon': df1['Lng_i'].mean()}),
                      title=dict(text="Evolución de los accidentes." , x=0.5,font=dict(color="ORANGE")),
                      font=dict(color="white"),
                    coloraxis_colorbar=dict(title=dict(text="Severidad", font=dict(color='white')))
                      )
    fig.update_coloraxes(colorbar=dict(
        bgcolor='black',  # Color de fondo de la barra temporal
        tickfont=dict(color='white'),  # Color de las etiquetas de la barra temporal
        title=dict(text='Tiempo', font=dict(color='white')),  # Título de la barra temporal
        title_font=dict(color='white')  # Color del título de la barra temporal
    ))

    # Cambiar la posición de los botones de reproducción y detención
    fig.update_layout(updatemenus=[
        dict(
            buttons=[
                dict(
                    args=[None, {"frame": {"duration": 500, "redraw": True}, "fromcurrent": True}],
                    label="Play",
                    method="animate",
                ),
                dict(
                    args=[[None], {"frame": {"duration": 0, "redraw": True}, "mode": "immediate", "transition": {"duration": 0}}],
                    label="Stop",
                    method="animate",
                )
            ],
            direction="left",  # Cambia la dirección de los botones a la izquierda
            pad={"r": 10, "t": 10},  # Espaciado de los botones
            showactive=False,
            x=-0.25,  # Posición horizontal de los botones
            xanchor="left",
            y=-0.48,  # Posición vertical de los botones
            yanchor="bottom"
        )
    ])

    return fig

#===================================callback6=============================================
@app.callback(
        [Output(component_id='graph_6', component_property='figure'),Output(component_id='graph_7', component_property='figure')],
        [Input(component_id='dropdown_6', component_property='value'),Input(component_id='checkllist', component_property='value')]
            )

def update_graph(dropdown_6,checkllist):

    df2 = df[['Provincia','Ciudad']].value_counts().reset_index()
    df2= df2.loc[df2['Provincia'] == checkllist]
    df2 = df2.rename(columns={0: 'count'})
    df2 = df2.sort_values(by='count', ascending=False)
    top5_ciudades = df2.head(5)

    fig2= px.pie(top5_ciudades , values='count', names='Ciudad',
                 color_discrete_sequence=px.colors.qualitative.Dark24
                 )
    fig2.update_layout(plot_bgcolor='black',paper_bgcolor='black',
                    title=dict(text="Porcentaje de ocurencia vs Ciudad." , x=0.5,font=dict(color="ORANGE")),
                    font=dict(color="white"),
                    coloraxis_colorbar=dict(title=dict(text="Severidad", font=dict(color='white')))
                        )
    #------------------------------------------------------------------------------------
    df3=df[['Provincia','Ciudad','Condicion_clima']]
    df3= df3.loc[df3['Provincia'] == checkllist]
    XX=top5_ciudades.Ciudad.unique()
    df3 = df3[df3['Ciudad'].isin(XX)]
    df4=df3[['Condicion_clima','Ciudad']].value_counts().reset_index()
    df4 = df4.rename(columns={0: 'count'})
    df4=df4[df4['count']>=20]
    df4= df4.sort_values(by='count', ascending=False)

    fig3= px.pie(df4 , values='count', names='Condicion_clima',
                 color_discrete_sequence=px.colors.qualitative.Bold
                 )
    fig3.update_layout(plot_bgcolor='black',paper_bgcolor='black',
                    title=dict(text="Porcentaje de ocurencia vs Clima." , x=0.5,font=dict(color="ORANGE")),
                    font=dict(color="white"),
                    coloraxis_colorbar=dict(title=dict(text="Severidad", font=dict(color='white')))
                        )
    return fig2,fig3
```
